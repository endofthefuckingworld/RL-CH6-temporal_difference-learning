```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
plt.style.use("seaborn")

GRID_SIZE = np.zeros((4,12))

ACTIONS = np.array([[-1, 0],   #0 up
                    [1, 0],    #1 down
                    [0, -1],   #2 left
                    [0, 1]])   #3 right

actions = np.arange(4)   #represent ACTIONS

STEP_SIZE = 0.5

DISCOUNT_RATE = 1

START = np.array([3,0])

TARGET = np.array([3,11])
class Environment:
    def __init__(self):
        self.state = START
        self.next_state = None
        self.action = np.random.choice(actions)
        
    def act_action(self,Action):
        pos = self.state + Action
        for i in range(len(pos)):
            if pos[i] < 0:
                pos[i] = max(0,pos[i])
            else:
                pos[i] = min(GRID_SIZE.shape[i]-1,pos[i])
                
        self.next_state = pos
        
        if pos[0] == 3 and 1 <= pos[1] <= 10: #fall off the cliff
            self.next_state = START
            return -100
        else:
            return -1
        
```

```python
def ε_greedy_policy(t,pos,Q_table):
    random_num = np.random.rand(1)
    if random_num > max(0.1,1/t):                #exploitation
        return np.argmax(Q_table[:,pos[0],pos[1]])
    else:                                        #exploration
        return np.random.choice(actions)
```

```python
def sarsa(t, q_table, expected = False, alpha = STEP_SIZE, gamma = DISCOUNT_RATE):
    actor = Environment()
    actor.action = ε_greedy_policy(t,actor.state,q_table)
    rewards = 0.0
    while (actor.state == TARGET).all() == False:
        reward = actor.act_action(ACTIONS[actor.action])
        rewards += reward
        next_action = ε_greedy_policy(t,actor.next_state,q_table)
        target = q_table[ next_action, actor.next_state[0], actor.next_state[1]]
        if expected:
            epsilon = max(0.1,1/t)
            target = 0.0
            q_next = q_table[:, actor.next_state[0], actor.next_state[1]]
            best_actions = np.argwhere(q_next == np.max(q_next))

            for action in actions:
                if action in best_actions:  # with probability[(1 - epsilon)/ len(best_action)] 
                    target += ((1 - epsilon) / len(best_actions) + epsilon / len(actions)) * \
                    q_table[ action, actor.next_state[0], actor.next_state[1]]
                else:                      # with probability(epsilon/ len(actions)) 
                    target += epsilon / len(actions) * \
                    q_table[ action, actor.next_state[0], actor.next_state[1]]

        q_table[actor.action,actor.state[0],actor.state[1]] += alpha * (
            reward + gamma * target - q_table[actor.action,actor.state[0],actor.state[1]]
        ) 
        
        actor.state = actor.next_state
        actor.action = next_action
        
    return rewards
    
```

```python
def q_learning(t, q_table, alpha = STEP_SIZE, gamma = DISCOUNT_RATE):
    actor = Environment()
    rewards = 0.0
    while (actor.state == TARGET).all() == False:
        actor.action = ε_greedy_policy(t,actor.state,q_table)
        reward = actor.act_action(ACTIONS[actor.action])
        rewards += reward
        
        q_table[actor.action,actor.state[0],actor.state[1]] += alpha * (
            reward + gamma * max(q_table[:,actor.next_state[0],actor.next_state[1]])
            -q_table[actor.action,actor.state[0],actor.state[1]]
        )
        
        actor.state = actor.next_state    
    return rewards
    
```

```python
def print_optimal_policy(Q_table):
    # display the optimal policy
    optimal_policy = []
    for i in range(0, 4):
        optimal_policy.append([])
        for j in range(0, 12):
            if ([i, j] == TARGET).all():
                optimal_policy[-1].append('G')
                continue
            bestAction = np.argmax(Q_table[:, i, j])
            if bestAction == 0:
                optimal_policy[-1].append('U')
            elif bestAction == 1:
                optimal_policy[-1].append('D')
            elif bestAction == 2:
                optimal_policy[-1].append('L')
            elif bestAction == 3:
                optimal_policy[-1].append('R')
    for row in optimal_policy:
        print(row)
```

```python
def training():
    # episodes of each run
    episodes = 250

    # perform 40 independent runs
    runs = 50

    rewards_sarsa = np.zeros(episodes)
    rewards_q_learning = np.zeros(episodes)
    rewards_e_sarsa = np.zeros(episodes)
    for r in tqdm(range(runs)):
        q_sarsa = np.zeros((4, 4, 12))
        q_e_sarsa = np.copy(q_sarsa)
        q_q_learning = np.copy(q_sarsa)
        for i in range(0, episodes):
            # cut off the value by -100 to draw the figure more elegantly
            # rewards_sarsa[i] += max(sarsa(q_sarsa), -100)
            # rewards_q_learning[i] += max(q_learning(q_q_learning), -100)
            rewards_sarsa[i] += sarsa(i+1, q_sarsa)
            rewards_e_sarsa[i] += sarsa(i+1, q_e_sarsa, True)
            rewards_q_learning[i] += q_learning(i+1, q_q_learning)
            
    rewards_sarsa /= runs
    rewards_e_sarsa /= runs
    rewards_q_learning /= runs
    
    return rewards_sarsa, rewards_e_sarsa,rewards_q_learning, q_sarsa, q_q_learning, q_e_sarsa
    
```

```python
rewards_sarsa, rewards_e_sarsa,rewards_q_learning, q_sarsa, q_q_learning, q_e_sarsa = training()
```

```python
print('Sarsa Optimal Policy:')
print_optimal_policy(q_sarsa)
print('Expected Sarsa Optimal Policy:')
print_optimal_policy(q_e_sarsa)
print('Q-Learning Optimal Policy:')
print_optimal_policy(q_q_learning)
```

```python
plt.plot(rewards_sarsa, label='Sarsa')
plt.plot(rewards_q_learning, label='Q-Learning')
plt.plot(rewards_e_sarsa, label='Expected-Sarsa')
plt.xlabel('Episodes')
plt.ylabel('Sum of rewards during episode')
plt.ylim([-100, 0])
plt.legend()
```
